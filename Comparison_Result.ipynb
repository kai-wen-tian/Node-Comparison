Repeated Transactions Multinode Percentage



```python
labels = ['Repeated Transactions Multinode', 'Unique Transactions Multinode']
sizes = [totalRepeatedTransactionsMultinode/totalTransactionsMultinode,uniqueTransactionsMultinode/totalTransactionsMultinode]  # These values represent the slice sizes


plt.pie(sizes, labels=labels,autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

Faster Transaction by Multinode

```python
data1 = {'Multinode': ((fasterTransactionMultinode/totalTransactionMultinodeOthers)*100), 'Others': (((totalTransactionMultinodeOthers-fasterTransactionMultinode)/totalTransactionMultinodeOthers)*100)}
data2 = {'Multinode': ((fasterTransactionMultinodeAlchemy/totalTransactionMultinodeAlchemy)*100), 'Alchemy': (((totalTransactionMultinodeAlchemy-fasterTransactionMultinodeAlchemy)/totalTransactionMultinodeAlchemy)*100)}
data3 = {'Multinode': ((fasterTransactionMultinodeInfura/totalTransactionMultinodeInfura)*100), 'Infura': (((totalTransactionMultinodeInfura-fasterTransactionMultinodeInfura)/totalTransactionMultinodeInfura)*100)}
data4={'Multinode': ((fasterTransactionMultinodeGoerli/totalTransactionMultinodeGoeri)*100), 'Goerli': (((totalTransactionMultinodeGoeri-fasterTransactionMultinodeGoerli)/totalTransactionMultinodeGoeri)*100)}

fig, axes = plt.subplots(2, 2, figsize=(10, 10))

axes[0, 0].bar(data1.keys(), data1.values(), color='blue')
axes[0, 0].set_title('Faster Transaction by Multinode')

axes[0, 1].bar(data2.keys(), data2.values(), color='green')
axes[0, 1].set_title('Faster Transction by Multinode vs Alchemy')

axes[1, 0].bar(data3.keys(), data3.values(), color='red')
axes[1, 0].set_title('Faster Transction by Multinode vs Infura')

axes[1, 1].bar(data4.keys(), data4.values(), color='purple')
axes[1, 1].set_title('Faster Transaction by Multinode vs Goerli')

plt.tight_layout()
plt.show()
```
